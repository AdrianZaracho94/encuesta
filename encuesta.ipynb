# ¿Cuál fue el motivo por el que no pudo concurrir a alguno/s de los talleres?

```python
motivo_inasistencia = df3[{'motivo_inasistencia'}]

print (motivo_inasistencia)
```

```python
(motivo_inasistencia
 .groupby("motivo_inasistencia")
 .agg(frecuencias=("motivo_inasistencia","count")))
```

# como vemos, la variable esta sucia. Hay que recategorizarla

```python
inasistencia_recode = motivo_inasistencia.replace({
'El horario no le convenía, No se enteró': 'No se enteró',
'El horario no le convenía, Tenía que cuidar a hijos/as/personas mayores o enfermos, Otros' : 'Tenía que cuidar a hijos/as/personas mayores o enfermos',
'El horario no le convenía, Tenía que cuidar a hijos/as/personas mayores o enfermos' : 'Tenía que cuidar a hijos/as/personas mayores o enfermos',
    
})
```

```python
# Vemos como quedo la recategorización

(inasistencia_recode
 .groupby("motivo_inasistencia")
 .agg(frecuencias=("motivo_inasistencia","count")))
```

```python
plt.figure(figsize=(10,5))
sns.countplot(x='motivo_inasistencia',data=inasistencia_recode)
plt.title('Motivo inasistencia a los talleres')
plt.xticks(rotation=45)
plt.show()
```
